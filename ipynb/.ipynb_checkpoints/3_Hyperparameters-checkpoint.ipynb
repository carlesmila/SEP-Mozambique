## Foto 2: Floor

```python
hyper_files = glob.glob('output/hyperparameters/foto2_hyper*.csv')
hyper_df = []
for i, fname in enumerate(hyper_files):
    df = pd.read_csv(fname, index_col=None, header=0)
    df['ID'] = str(i)
    hyper_df.append(df)

del(df)
hyper_df = pd.concat(hyper_df, axis=0, ignore_index=True)
```

```python
hyper_df = pd.melt(hyper_df, id_vars=['epoch', 'ID'], var_name='var')
hyper_df['Dataset'] = ['Train' if ('train' in i) else 'Test' for i in hyper_df['var']]
hyper_df['Metric'] = ['Loss' if ('loss' in i) else 'Accuracy assets' if ('assets_acc' in i) else 'Accuracy expenditure' if ('exp_acc' in i) else 'Accuracy income' 
                             for i in hyper_df['var']]
hyper_df = hyper_df.drop('var', axis=1)
hyper_df = hyper_df.loc[hyper_df.ID!='6'] # delete #6 it is so bad it distorts loss function scales
```

```python
# Results train data
g = sns.FacetGrid(hyper_df.loc[hyper_df.Dataset=='Train'], col="Metric", col_wrap=2, legend_out=True, height=3, aspect=1.5, sharey=False)
g.set(xticks=np.arange(0,51,5))
g.map_dataframe(sns.lineplot, x="epoch", y="value", hue="ID")
g.add_legend()
g.legend.set_title("Hyper.\nconfig.")
```

```python
# Results test data
g = sns.FacetGrid(hyper_df.loc[hyper_df.Dataset=='Test'], col="Metric", col_wrap=2, legend_out=True, height=3, aspect=1.5, sharey=False)
g.set(xticks=np.arange(0,51,5))
g.map_dataframe(sns.lineplot, x="epoch", y="value", hue="ID")
g.add_legend()
g.legend.set_title("Hyper.\nconfig.")
```
